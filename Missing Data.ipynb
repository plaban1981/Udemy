# Duplicate Data

```python
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
url = 'http://bit.ly/movieusers'
users = pd.read_table(url, sep='|', header=None, names=user_cols, index_col='user_id')
```

```python
users.shape
```

```python
users.info()
```

```python
users.head()
```

```python
#This code lists out the duplicated zip codes in our dataframe.
users.zip_code.duplicated()
```

```python
#The code in this step gives us the number of zip codes that are duplicated in our dataframe
users.zip_code.duplicated().sum()
```

```python
users.loc[users.duplicated(), :]
```

```python
#In this code we are marking the duplicates to true except for the first occurrence. 
users.loc[users.duplicated(keep='first'), :]
```

```python
users.loc[users.duplicated(keep='last'), :]
```

```python
#In the code below we use False to drop all of the duplicates. 
users.loc[users.duplicated(keep=False), :]
```
